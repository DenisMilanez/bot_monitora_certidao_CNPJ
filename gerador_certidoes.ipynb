```python
import pandas as pd

df_clientes = pd.read_excel('to_ignore/Pasta3.xlsx', sheet_name='config_clientes', index_col=0)
df_cert_config = pd.read_excel('to_ignore/Pasta3.xlsx', sheet_name='config_certidoes', index_col=0)

# Passo 1: Explode a coluna CERTIDOES_CADASTRADAS em múltiplas linhas
df_clientes_expanded = df_clientes.assign(
    CERTIDAO_ID=df_clientes['CERTIDOES_CADASTRADAS'].str.split(', ')
).explode('CERTIDAO_ID')

# Passo 2: Faz o merge entre df_clientes_expanded e df_cert_config
df_merged = df_clientes_expanded.merge(
    df_cert_config, 
    left_on='CERTIDAO_ID', 
    right_index=True, 
    how='left'
)

# Passo 3: Seleciona as colunas desejadas no resultado final
df_resultado = df_merged[[
    'NOME_CLIENTE', 'CNPJ', 'NOME_COMPLETO', 'NOME_ARQUIVO', 'PASTA_DESTINO', 
    'CERTIDAO_ID', 'TITULO_CERTIDAO', 'ORGAO', 'URL'
]]

# Exibe o DataFrame final
df_resultado.info()
df_resultado.head(100)
```

```python
import pandas as pd
import random
from fpdf import FPDF
from datetime import datetime, timedelta
import os  # Importando a biblioteca os
from babel.dates import format_date


# Definindo as datas de início e validade
data_inicio_min = datetime(2024, 9, 1)
data_inicio_max = datetime(2024, 10, 5)
data_validade_min = datetime(2024, 11, 10)
data_validade_max = datetime(2024, 11, 17)


def gerar_data_aleatoria(inicio, fim):
    return inicio + timedelta(days=random.randint(0, (fim - inicio).days))

# Função para gerar um número de certificado aleatório
def gerar_numero_certificado():
    return ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=10))

# Função para criar PDF
def criar_pdf(certidao_id, orgao, titulo_certidao, numero_certificado, data_inicio, nome_completo, cnpj, data_validade, url, pasta_destino, nome_arquivo):
    pdf = FPDF()
    pdf.add_page()

    # Definindo fonte
    pdf.set_font("Arial", size=12)

    # Adicionando texto ao PDF
    pdf.cell(200, 10, f"Orgão: {orgao}", ln=True)
    pdf.cell(200, 10, titulo_certidao, ln=True)

    if certidao_id == 'CEF-CRFGTS':
        pdf.cell(200, 10, f"Certificação Número: {numero_certificado}", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.multi_cell(0, 10, f"CERTIFICAMOS que, após consulta aos registros eletrônicos disponíveis, NADA CONSTA contra o nome por extenso e CPF/CNPJ de:\n{nome_completo}\n{cnpj}")
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Validade:{data_inicio.strftime('%d/%m/%Y')} a {data_validade.strftime('%d/%m/%Y')}", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Informação obtida em :{data_inicio.strftime('%d/%m/%Y')} 10:07:23", ln=True)
    
    elif certidao_id == 'TJDFT-CNFR':
        pdf.multi_cell(0, 10, f"Certifico que nesta data ({data_inicio.strftime('%d/%m/%Y')} às 06:53) NÃO CONSTA no Cadastro CERTIFICAMOS que, após consulta aos registros eletrônicos disponíveis até {data_inicio.strftime('%d/%m/%Y')}, NADA CONSTA contra o nome por extenso e CPF/CNPJ de:\n{nome_completo}\n{cnpj}")
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Emitida gratuitamente pela internet em: {data_inicio.strftime('%d/%m/%Y')}", ln=True)   
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Selo digital de segurança: {numero_certificado}", ln=True)
    
    elif certidao_id == 'CNJ-CNIADM':
        pdf.multi_cell(0, 10, f"CERTIFICO que nesta data ({data_inicio.strftime('%d/%m/%Y')} às 06:53), NADA CONSTA contra o nome por extenso e CPF/CNPJ de:\n{nome_completo}\n{cnpj}")
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.multi_cell(200, 10, f"""Esta certidão é expedida gratuitamente. Sua autenticidade pode ser por meio do número de controle\n672D.DF85.D1C3.5589 no seguinte endereço: https://www.cnj.jus.br/improbidade_adm/autenticar_certidao.php""")
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Gerado em: {data_inicio.strftime('%d/%m/%Y')} as 06:53:0m", ln=True)

    elif certidao_id == 'CGU-CNC':
        pdf.multi_cell(0, 10, f"Certifica-se que, em consulta aos sistemas ePAD e CGU-PJ e aos cadastros CEIS, CNEP e CEPIM mantidos pela Corregedoria-Geral da União, NÃO CONSTAM registros de penalidades vigentes relativas ao CNPJ/CPF consultado:{cnpj}")
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Certidão emitida às 10:11:43 do dia {data_inicio.strftime('%d/%m/%Y')} , com validade até o dia {data_validade.strftime('%d/%m/%Y')}.", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Código de controle da certidão: {numero_certificado}", ln=True)

    elif certidao_id == 'FN-CNDTF':
        pdf.multi_cell(0, 10, f"é certificado que não constam pendências em seu nome, relativas a créditos tributários administrados pela Secretariada Receita Federal do Brasil consultado:{cnpj}")
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Emitida às 14:17:48 do dia {data_inicio.strftime('%d/%m/%Y')} <hora e data de Brasília>.", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Válida até {data_validade.strftime('%d/%m/%Y')}.", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Código de controle da certidão: {numero_certificado}", ln=True)

    elif certidao_id == 'TST-CNDT':
        pdf.cell(200, 10, f"Certidão nº: {numero_certificado}", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Expedição: {data_inicio.strftime('%d/%m/%Y')}, às 10:25:22", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Validade: {data_validade.strftime('%d/%m/%Y')}.", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.multi_cell(0, 10, f"CERTIFICA-SE QUE {nome_completo}, inscrito no CNPJ n. {cnpj}, NÃO CONSTA como inadimplente ao Banco Nacional de Devedores Trabalhistas")

    elif certidao_id == 'GDF-DIF':
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Este documento foi emitido no dia {data_inicio.strftime('%d/%m/%Y')} na Internet pelo portal Agênci@Net", ln=True)

    elif certidao_id == 'GDF-CND':
        pdf.cell(200, 10, f"CERTIDÃO Nº:{numero_certificado}", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.multi_cell(0, 10, f"Até esta data não constam débitos de tributos de competência do Distrito Federal, inclusive os relativos à Divida Ativa, referentes a {nome_completo}, inscrito no CNPJ n. {cnpj}")
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        data_formatada = format_date(data_validade, format='dd \'de\' MMMM \'de\' y', locale='pt_BR')
        pdf.cell(200, 10, f"Válida até {data_formatada}.", ln=True)
        pdf.cell(0, 10, '', ln=True)  # Linha em branco
        pdf.cell(200, 10, f"Certidão emitida via internet em {data_inicio.strftime('%d/%m/%Y')} às 10:20:37 e deve ser validada no endereço https://www.receita.fazenda.df.gov.br", ln=True)
    
    pdf.cell(0, 10, '', ln=True)  # Linha em branco
    pdf.cell(200, 10, f"Certidão disponível em: {url}", ln=True)

    # Criando o diretório se não existir
    if not os.path.exists(pasta_destino):
        os.makedirs(pasta_destino)

    arquivo_nome = f"{data_inicio.strftime('%Y-%m-%d')}_{nome_arquivo}.pdf"
    pdf.output(os.path.join(pasta_destino, arquivo_nome))  # Usando os.path.join para criar o caminho


# Iterando sobre cada linha do DataFrame
for index, row in df_resultado.iterrows():
    data_inicio = gerar_data_aleatoria(data_inicio_min, data_inicio_max)
    data_validade = gerar_data_aleatoria(data_validade_min, data_validade_max)
    numero_certificado = gerar_numero_certificado()

    criar_pdf(certidao_id=row['CERTIDAO_ID'],
        orgao=row['ORGAO'],
        titulo_certidao=row['TITULO_CERTIDAO'],
        numero_certificado=numero_certificado,
        data_inicio=data_inicio,
        nome_completo=row['NOME_COMPLETO'],
        cnpj=row['CNPJ'],
        data_validade=data_validade,
        url=row['URL'],  # Corrigido para estar na posição correta
        pasta_destino=row['PASTA_DESTINO'],
        nome_arquivo=row['NOME_ARQUIVO']
    )
```
